1. Print all numbers from 1 to 10 using a while loop.


```python
num = 1
while num <= 10:
    print(num)
    num += 1
```

2. Write a program that asks the user to input a number and keeps asking for input until they enter a number greater than 10.


```python
num = int(input("Enter a number greater than 10: "))
while num <= 10:
    print("Try again!")
    num = int(input("Enter a number greater than 10: "))
print(f"Good job! You entered {num}.")
```

3. Write a program to find the sum of all positive integers less than or equal to a given number n.


```python
n = int(input("Enter a number: "))
total = 0
i = 1
while i <= n:
    total += i
    i += 1
print(f"The sum of numbers from 1 to {n} is {total}.")
```

4. Create a while loop that counts the number of vowels in a given string.


```python
string = input("Enter a string: ")
vowels = "aeiouAEIOU"
count = 0
index = 0
while index < len(string):
    if string[index] in vowels:
        count += 1
    index += 1
print(f"The number of vowels in the string is {count}.")
```

5. Write a program that asks the user to guess a secret number (between 1 and 100) until they guess correctly. The program should give hints (too low or too high) after each guess.


```python
import random

secret_number = random.randint(1, 100)
guess = 0
while guess != secret_number:
    guess = int(input("Guess the secret number (1-100): "))
    if guess < secret_number:
        print("Too low!")
    elif guess > secret_number:
        print("Too high!")
print("Congratulations! You guessed the number correctly.")
```

6. Write a program that prints the Fibonacci sequence up to the nth term, where n is provided by the user.

```python
n = int(input("Enter the number of terms for the Fibonacci sequence: "))
a, b = 0, 1
count = 0
while count < n:
    print(a, end=" ")
    a, b = b, a + b
    count += 1
```

7. Write a program that prints all the odd numbers between 1 and 20 using a while loop.


```python
num = 1
while num <= 20:
    if num % 2 != 0:
        print(num)
    num += 1
```

8. Write a program that repeatedly asks the user for their name and age until they enter a valid age (greater than 0).


```python
while True:
    name = input("Enter your name: ")
    age = int(input("Enter your age: "))
    if age > 0:
        print(f"Hello {name}, you are {age} years old.")
        break
    else:
        print("Please enter a valid age (greater than 0).")
```

9. Write a program that calculates the factorial of a number entered by the user using a while loop.


```python
n = int(input("Enter a number to calculate its factorial: "))
factorial = 1
i = 1
while i <= n:
    factorial *= i
```

10. Write a program that finds the greatest common divisor (GCD) of two numbers using a while loop.


```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
print(f"The GCD of {num1} and {num2} is {gcd(num1, num2)}.")
```
